**Part A – Class Creation**

●items → List of item prices (list of numbers)

●Use init() to initialize all attributes correctly

**Part B – Core Methods Implementation**  
Implement the following methods inside the ShoppingCart class:  

**1. add_item(price)**  
●Adds a new item price to the cart

●Price must be a positive number  


**2. remove_item(price)**  
●Removes one occurrence of the given price from the cart  

●If the item does not exist, display a suitable message  


**3. calculate_total()**    
●Uses a loop or list comprehension  

●Returns the total price of all items in the cart  


**4. apply_discount()**  
●Discount rule:  
➤Total ≥ 3000 → apply 10% discount  
➤Total < 3000 → no discount  

●Returns the final payable amount

```python
class ShoppingCart:

    def __init__(self, customer_name, items=None):

      # Initialize customer name
      self.customer_name = customer_name

      # If no items provided, use an empty list
      if items is None:
        self.items = []
      else:
        self.items = items

    # 1. Add Item
    def add_item(self, price):
      if price > 0:
          self.items.append(price)
      else:
          print("Invalid price! Price must be a positive number.")

    # 2. Remove Item
    def remove_item(self, price):
      if price in self.items:
          self.items.remove(price)  # removes one occurrence
      else:
          print("Item not found in the cart.")

    # 3. Calculate Total
    def calculate_total(self):
      return sum(self.items)   # using list aggregation

    # 4. Apply Discount
    def apply_discount(self):
      total = self.calculate_total()

      if total >= 3000:
          discount = total * 0.10
          final_amount = total - discount
      else:
          final_amount = total

      return final_amount
```

**Part C – Display & Utility Methods**  
**1. display_cart()**  

Prints the following in a well-formatted manner:
●Customer name  
●Items in the cart  
●Total amount  
●Final amount after discount  
●Number of items in the cart  

**2. clear_cart()**  
●Removes all items from the cart  
●Displays a confirmation message

```python
class ShoppingCart:

    def __init__(self, customer_name, items=None):

      # Initialize customer name
      self.customer_name = customer_name

      # If no items provided, use an empty list
      if items is None:
        self.items = []
      else:
        self.items = items

    # Add Item
    def add_item(self, price):
      if price > 0:
          self.items.append(price)
      else:
          print("Invalid price! Price must be a positive number.")

    # Remove Item
    def remove_item(self, price):
      if price in self.items:
          self.items.remove(price)  # removes one occurrence
      else:
          print("Item not found in the cart.")

    # Calculate Total
    def calculate_total(self):
      return sum(self.items)   # using list aggregation

    # Apply Discount
    def apply_discount(self):
      total = self.calculate_total()

      if total >= 3000:
          discount = total * 0.10
          final_amount = total - discount
      else:
          final_amount = total

      return final_amount

    # Display Cart
    def display_cart(self):
      total = self.calculate_total()
      final_amount = self.apply_discount()

      print("\n------------- Shopping Cart Details ------------")
      print(f"Customer Name                : {self.customer_name}")
      print(f"Items                        : {self.items}")
      print(f"Total Amount                 : {total}")
      print(f"Final Amount after discount  : {final_amount}")
      print(f"Total Items                  : {len(self.items)}")
      print("--------------------------------------------------")


    # Clear Cart
    def clear_cart(self):
      self.items.clear()
      print("Cart has been cleared successfully.")
```
