**Question 1: Enumerate with List**  
Given the list:

fruits = ["apple", "banana", "cherry", "mango"]

Use enumerate() to print each fruit with its index number.

Output format example:

0 apple  

1 banana  

2 cherry  

3 mango

```python
fruits = ["apple", "banana", "cherry", "mango"]

for index, value in enumerate(fruits):
  print(index, value)
```

**Question 2: Zip with Two Lists**  
Given two lists:

students = ["Rahim", "Karim", "Ayesha"]

scores = [85, 90, 78]

Use zip() to print each student name along with their score.

Example Output:

Rahim scored 85  

Karim scored 90  

Ayesha scored 78

```python
students = ["Rahim", "Karim", "Ayesha"]

scores = [85, 90, 78]

for student, score in zip(students, scores):
  print(f"{student} scored {score}")
```

**Question 3: Tuple and Loop**  
Create a tuple of 5 subjects.

Using a for loop, print each subject one by one.

```python
subjects = ("Discrete Math", "Algorithm", "Data Structure", "Electrical Circuit", "System Design")

for subject in subjects:
    print(subject)
```

**Question 4: Set Operations**  
Given the sets:

set1 = {1, 2, 3, 4, 5}

set2 = {4, 5, 6, 7}

Print the common elements using set operation

Print the unique elements of set1

```python
set1 = {1, 2, 3, 4, 5}

set2 = {4, 5, 6, 7}

common = set1 & set2
print(f"Common elements are: {common}")

unique_elements = set1 - set2
print(f"Unique elements of set1: {unique_elements}")
```
